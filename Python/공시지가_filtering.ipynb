# Load Data

```python
with open(f'./processed_data.pkl', 'rb') as f:
    data_dict = pickle.load(f)
```

```python
land_money_shp = gpd.read_file("./서울_공시지가_shp/AL_D150_11_20240125.shp")
```

```python
land_money_shp.crs = "EPSG:5186"
land_money_shp = land_money_shp.to_crs(epsg=5181)
```

# Filtering

```python
filtered_land_money_shp = gpd.sjoin(data_dict["gwangjin_admin"], land_money_shp, how="inner", op='intersects')
```
